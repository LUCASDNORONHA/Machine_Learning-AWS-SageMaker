# Pr√©-processamento

```python
dfCasas.columns
```

```python
X = dfCasas.iloc[:, 3:19].values
y = dfCasas.iloc[:, 2].values
```

```python
X.shape, y.shape
```

```python
type(X), type(y)
```

```python
X = np.array(X).astype('float32')
y = np.array(y).astype('float32')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state= 0)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

# Configura√ß√µes SageMaker

```python
import sagemaker
import boto3
from sagemaker import Session
```

```python
session = sagemaker.Session()

name_bucket = 'awsmysagemakers3'
subpasta_modelo = 'modelos/house-prices/linear-learner'
subpasta_dataset = 'datasets/house-prices'
```

```python
role = sagemaker.get_execution_role()
```

```python
import io
import sagemaker.amazon.common as smac

buffer = io.BytesIO()
smac.write_numpy_to_dense_tensor(buffer, X_train, y_train)
buffer.seek(0)
```

```python
import os

key = 'house-train-data'
boto3.resource('s3').Bucket(name_bucket).Object(os.path.join(subpasta_dataset, 'train', key)).upload_fileobj(buffer)
```

```python
s3_train_data = f's3://{name_bucket}/{subpasta_dataset}/train/{key}'

print(f'Localiza√ß√£o da base de treinamento: {s3_train_data}')
```

```python
output_location = f's3://{name_bucket}/{subpasta_modelo}/output'

print(f'Modelo final ser√° salvo em: {output_location}')
```

# Cria√ß√£o e Treinamento do Linear Learner

```python
container = sagemaker.image_uris.retrieve(framework = 'linear-learner', region=boto3.Session().region_name)
```

```python
linear = sagemaker.estimator.Estimator(image_uri= container, 
                                       role= role, 
                                       instance_count= 1, 
                                       instance_type= 'ml.m4.xlarge', 
                                       output_path= output_location,
                                      sagemaker_session= session)
```

```python
linear.set_hyperparameters(predictor_type='regressor')
```

```python
linear.fit({'train': s3_train_data})
```

# Deploy, Previs√µes e Avalia√ß√£o

```python
linear_regressor = linear.deploy(initial_instance_count=1, instance_type='ml.m4.xlarge')
```

```python
from sagemaker.serializers import CSVSerializer
from sagemaker.deserializers import JSONDeserializer
```

```python
linear_regressor.serializer = CSVSerializer()
linear_regressor.deserializer = JSONDeserializer()
```

```python
results = linear_regressor.predict(X_test)
```

```python
results
```

```python
previsoes = np.array([r['score'] for r in results['predictions']])
```

```python
previsoes
```

```python
previsoes.shape, X_test.shape
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae = mean_absolute_error(y_test, previsoes)
mse = mean_squared_error(y_test, previsoes)
print('MAE = ', mae, 'MSE = ', mse)
```

```python
linear_regressor.delete_endpoint()
```
